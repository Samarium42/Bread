```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Small sample size for ease of analysis
df = pd.read_excel('Superstore Dataset.xlsx').sort_values(by='Order Date')
df = df.head(1000)
```

```python
# Make dataframe summing sales across a given subcategory per day
categorical = df.groupby(['Sub-Category', 'Order Date'])['Sales'].sum()
categorical = pd.DataFrame(categorical).reset_index()

# Lets look at the sales of accessories
accessories_sales = categorical[categorical['Sub-Category']=='Accessories'].sort_values(by='Order Date')
```

```python
# Make dataframe getting the average discount for that subcategory per day
category_discount = df.groupby(['Sub-Category', 'Order Date'])['Discount'].mean()
category_discount = pd.DataFrame(category_discount).reset_index()
# Looking at accessories
accessories = category_discount[category_discount['Sub-Category'] == 'Accessories'].sort_values(by='Order Date')
```

```python
accessories_sales['Discount'] = accessories['Discount']
accessories_sales
# Plot discount and sales with discount scaled for visibility 
plt.figure(figsize=(50,10))
plt.plot(accessories_sales['Order Date'], accessories_sales['Discount'] * 1000)
plt.plot(accessories_sales['Order Date'], accessories_sales['Sales'])
plt.plot()
```

From this analysis there does not appear to be a direct correlation between sales and discount. In some regards it seems as though sales are at their highest with no discount
